# Multi Layer Perceptron Architecture and Forward Propagation

## Perceptron by Rosenblatt

A simple computational unit called the **perceptron**, which was the foundational building block for the large neural networks we use today.

In 1958, Frank Rosenblatt introduced the perceptron, a pioneering model of artificial neural networks.

He envisioned it as a machine capable of
- learning from its environment,
- making decisions, and
- potentially performing tasks such as language translation.

Rosenblatt's early work laid the foundation for the field of machine learning and artificial intelligence.

In his 1958 proposal for funding the "Cornell Photoperceptron," Rosenblatt claimed that devices of this sort were expected to be capable of concept formation, language translation, collation of military intelligence, and the solution of problems through inductive logic. This ambitious vision highlighted the potential of the perceptron to perform complex cognitive tasks.
Rosenblatt's work on the perceptron was an early step toward creating machines that could mimic certain aspects of human cognition.


A perceptron is fundamentally designed as a binary classifier that determines whether an input belongs to one class or another. To understand how it works, imagine a simple binary classification task where we need to decide if an input (such as an image, sound, or data point) belongs to category A or category B.

### How a perceptron work?

The perceptron computes a weighted sum of multiple inputs, adds a bias term, and then applies a step function to this cumulative input.

- The cumulative input to the perceptron is calculated as the sum of each weight multiplied by its corresponding input plus a bias term:

$$
\text{Cumulative Input} = w_1 x_1 + w_2 x_2 + \cdots + w_k x_k + b
$$

- Specifically, it outputs **1** if the cumulative input is positive and **0** otherwise. In other words, the perceptron "fires" (outputs 1) when the total input exceeds zero and remains "inactive" (outputs 0) when it does not.

- The step function can be defined in different ways. Here, we use:

$$
y =
\begin{cases}
1 & \text{if } x \geq 0 \\
0 & \text{if } x < 0
\end{cases}
$$

**Note: Some definitions use 1 and -1 instead of 1 and 0.**

#### Vector Representation

To work more efficiently, we use vector and matrix notation. Let **$ \mathbf{w} $** and **$ \mathbf{x} $** be vectors representing the weights and inputs respectively (both assumed as column vectors):

$$
\mathbf{w} =
\begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_k
\end{bmatrix}
\quad
\mathbf{x} =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_k
\end{bmatrix}
$$

The weighted sum of inputs can be neatly expressed as the dot product between the transpose of the weights vector and the inputs vector:

$$
\mathbf{w}^T \mathbf{x} = [w_1, w_2, \ldots, w_k] \cdot
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_k
\end{bmatrix}
= w_1 x_1 + w_2 x_2 + \cdots + w_k x_k
$$

Adding the bias term, the cumulative input becomes:

$$
\mathbf{w}^T \mathbf{x} + b = w_1 x_1 + w_2 x_2 + \cdots + w_k x_k + b
$$

**Final Output**

Applying the step function to this cumulative input, the perceptron outputs:

$$
y =
\begin{cases}
1 \quad & \text{if } \mathbf{w}^T \mathbf{x} + b \geq 0 \\
0 \quad & \text{otherwise}
\end{cases}
$$

This simple mechanism allows the perceptron to act as a binary classifier, firing when the weighted combination of inputs crosses a threshold, making it a fundamental building block for neural network architectures.


**Visual Explanation**
- **Input Layer:** Features of the object.
- **Weighted Sum:** Each features contribution is scaled by its weight.
- **Activation:** If the sum exceeds a threshold, the perceptron classifies the input as positive; otherwise, negative.


The key objective in training a perceptron for binary classification is to find the optimal weight vector **$ \mathbf{w} $** and bias **$ b $** such that the linear decision function correctly separates the two classes:

$$
\mathbf{w}^T \mathbf{x} + b \geq 0 \quad \text{for all examples where } y = 1
$$

and

$$
\mathbf{w}^T \mathbf{x} + b < 0 \quad \text{for all examples where } y = 0
$$

Here, the perceptron uses the following step function as its activation:

$$
y =
\begin{cases}
1 & \text{if } z \geq 0 \\
-1 & \text{if } z < 0
\end{cases}
\quad \text{where } z = \mathbf{w}^T \mathbf{x} + b
$$

This formulation means the perceptron outputs a positive class label (1) when the weighted sum of inputs plus bias crosses the threshold zero, and outputs the negative class (0) otherwise.


Using uniform coordinates, the problem of training a perceptron to separate two classes can be expressed more compactly. By augmenting the input vector $\mathbf{x}$ with the bias term $b$, we define an extended input vector $\tilde{\mathbf{x}}$:

$$
\tilde{\mathbf{x}} = \begin{bmatrix} \mathbf{x} \\ 1 \end{bmatrix}
$$

and similarly extend the weight vector to include the bias:

$$
\tilde{\mathbf{w}} = \begin{bmatrix} \mathbf{w} \\ b \end{bmatrix}
$$

The classification condition then becomes a simple dot product:

$$
\tilde{\mathbf{w}}^T \tilde{\mathbf{x}} \geq 0 \quad \text{for all examples where} \quad y = 1
$$

and

$$
\tilde{\mathbf{w}}^T \tilde{\mathbf{x}} < 0 \quad \text{for all examples where} \quad y = 0
$$

The perceptron activation is defined by the step function:

$$
y =
\begin{cases}
+1 & \text{if } \tilde{\mathbf{w}}^T \tilde{\mathbf{x}} \geq 0 \\
-1 & \text{if } \tilde{\mathbf{w}}^T \tilde{\mathbf{x}} < 0
\end{cases}
$$

This formulation unifies the weight and bias into a single vector, simplifying both notation and computation, and makes the perceptron's decision a direct function of the dot product between the augmented weight and input vectors.




#### Geometric Intuition



The perceptron seeks a **linear decision boundary** in $$ \mathbb{R}^n $$ defined by the equation:
$$
\mathbf{w}^T \mathbf{x} + b = 0
$$
- Points for which $$ \mathbf{w}^T \mathbf{x} \geq 0 $$ belong to the *positive class*.
- Points for which $$ \mathbf{w}^T \mathbf{x} < 0 $$ belong to the *negative class*.

The weight vector $ \mathbf{w} $ is **perpendicular** to the decision boundary. The bias $ b $ shifts this boundary without changing its orientation.

  - If the angle between $\mathbf{w}$ and $\mathbf{x}$ is **less than 90**, their dot product is positive, indicating a positive class.
  - If the angle is **greater than 90**, the dot product is negative, indicating a negative class.

- Thus, learning involves adjusting $\mathbf{w}$ and $b$ so that this hyperplane correctly separates the classes in the feature space.

  - For any pair of vectors $ \mathbf{w} $ and $ \mathbf{x} $, the dot product is given by:
  $$
  \mathbf{w}^T \mathbf{x} = \|\mathbf{w}\| \|\mathbf{x}\| \cos(\theta)
  $$
  where $ \theta $ is the **angle** between $ \mathbf{w} $ and $ \mathbf{x} $.

    - If $ \mathbf{w}^T \mathbf{x} > 0 $, then $ \cos(\theta) > 0 \Rightarrow \theta < 90^\circ $: the vector $ \mathbf{x} $ lies **in the same half-space** as the positive side of $ \mathbf{w} $.
    - If $ \mathbf{w}^T \mathbf{x} < 0 $, then $ \cos(\theta) < 0 \Rightarrow \theta > 90^\circ $: the vector $ \mathbf{x} $ lies on the **opposite side**.

  - Therefore, classification in a perceptron depends entirely on the **sign** of the dot product.

#### Why the Update Works?  Geometric Explanation

Consider two possible misclassifications:

(a) **Positive Example Misclassified as Negative:**
For a positive sample $ \mathbf{x}_P $:

$$
\mathbf{w}^T \mathbf{x}_P < 0 \quad \text{(wrong: should be positive)}
$$
The correction step is:
$$
\mathbf{w}_{\text{new}} = \mathbf{w}_{\text{old}} + \eta \mathbf{x}_P
$$

**Result:** The term $ \mathbf{w}_{\text{new}}^T \mathbf{x}_P $ increases because we are adding a component in the direction of $ \mathbf{x}_P $, effectively reducing the angle $ \theta $ between $ \mathbf{w} $ and $ \mathbf{x}_P $ (making $ \cos\theta $ larger). Thus, the positive point is pushed to the correct side.

(b) **Negative Example Misclassified as Positive**
For a negative sample $ \mathbf{x}_N $:
$$
\mathbf{w}^T \mathbf{x}_N > 0 \quad \text{(wrong: should be negative)}
$$
The correction step is:
$$
\mathbf{w}_{\text{new}} = \mathbf{w}_{\text{old}} - \eta \mathbf{x}_N
$$
**Result:** The subtraction moves $ \mathbf{w} $ away from $ \mathbf{x}_N $, increasing their angle beyond $ 90^\circ $, so $ \mathbf{w}^T \mathbf{x}_N < 0 $.


##### Summary

- The perceptron iteratively rotates the weight vector to reduce the angle between $ \mathbf{w} $ and correctly classified samples.
- Positive examples push $ \mathbf{w} $ toward them; negative ones repel $ \mathbf{w} $ away.
- Each update improves alignment until all points are correctly classified.

Hence, **the perceptron learning algorithm works because each weight update geometrically moves the decision boundary closer to an orientation that correctly separates the data**, guaranteeing convergence when the data are linearly separable.

$$
\boxed{
\begin{aligned}
\mathbf{w}_{\text{new}} &= \mathbf{w}_{\text{old}} + \eta y_i \mathbf{x}_i \\
\mathbf{w}_{\text{new}}^T \mathbf{x}_i &> \mathbf{w}_{\text{old}}^T \mathbf{x}_i \text{ if misclassified positive sample} \\
\mathbf{w}_{\text{new}}^T \mathbf{x}_i &< \mathbf{w}_{\text{old}}^T \mathbf{x}_i \text{ if misclassified negative sample}
\end{aligned}}
$$
Through this update mechanism, the angle between $ \mathbf{w} $ and correctly labeled examples is iteratively adjusted until the perceptron finds the correct linear separator.

#### Perceptron Learning Algorithm Overview

**An iterative solution suggested by Rosenblatt for training the perceptron.**

1. **Initialize** the weight vector $\tilde{\mathbf{w}}$ (including bias) with small random values or zeros.

2. **For each training sample** $(\tilde{\mathbf{x}}, y)$ where $y \in \{1, 0\}$:

   - Compute the predicted label:
     
     $$
     \hat{y} = \text{sign}(\tilde{\mathbf{w}}^T \tilde{\mathbf{x}})
     $$
   
   - If the prediction $\hat{y}$ does **not** match the true label $y$ (i.e., the sample is misclassified), perform the weight update:
     
     $$
     \tilde{\mathbf{w}} \leftarrow \tilde{\mathbf{w}} + \eta \. y \. \tilde{\mathbf{x}}
     $$

      - $\tilde{\mathbf{x}}$ is a misclassified data point (in homogeneous coordinates),
      - $y$ is its true label, and
      - the expression $y \. \tilde{\mathbf{x}}$ represents element-wise scaling, **not a dot product**.
   
   Here, $\eta > 0$ is the learning rate controlling the size of the update step.

3. **Repeat** the above process iteratively for the entire training set until either all points are correctly classified or a maximum number of iterations is reached.


  - The perceptron adjusts $\tilde{\mathbf{w}}$ in the direction of the misclassified inputs scaled by their labels, effectively moving the decision boundary closer to a state that separates the classes correctly.

  - Representing input and weights in uniform coordinates simplifies the model by incorporating the bias directly into the weight vector, making all operations rely on simple dot products.


**Example:**
Suppose you want to classify emails as spam or not spam based on features like the presence of certain keywords. The perceptron learns the optimal weights during training, eventually firing (outputting 1) when the weighted combination of features indicates spam, and staying inactive (outputting 0) otherwise.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 72
from numpy import array, sin, cos, dot, pi, arange, linspace, meshgrid
from ipywidgets.widgets import interactive, interact, interactive_output, HBox, VBox, FloatSlider, Label, Text, jslink

# Hyperplane xw+b=0
def hyper(x1, w, b):
    # x1, x2 are scalar/vector, w is vector, b is scalar
    # Find and return x2 from x1 that is on the hyperplane, i.e. solve the equation wx+b=0
    return -(x1*w[0]+b)/w[1] if w[1] != 0 else 0

# Classifier
def perceptron(x, w, b):
    # x, w are vectors, b is scalar
    # Return the classification, i.e. f(x)
    return 1 if dot(x,w)+b > 0 else 0
```

```python
# h controls the display granularity
h = 1.7

# Display x1 and x2 limits
x1_min, x1_max = -10, 10
x2_min, x2_max = -10, 10

# Widgets in decorator
@interact(theta=FloatSlider(value=pi/8, min=-pi/1.7, max=pi/1.7), b=(-7.,7.))
def plot_w(theta, b):
    plt.figure(figsize=(8, 8), dpi=72)

    # theta controls the angle of the hyperplane
    w = array([cos(theta), sin(theta)])

    # Plot normal arrows
    x1 = linspace(2*sin(theta)*x1_min, 2*sin(theta)*x1_max, 11)
    x2 = hyper(x1, w, b)
    x1_end = x1 + h*w[0]
    x2_end = x2 + h*w[1]
    for x1_o,x2_o,x1_e,x2_e in zip(x1,x2,x1_end,x2_end):
        plt.arrow(x1_o, x2_o, x1_e-x1_o, x2_e-x2_o, head_width=.3, color='r')

    # Plot the hyperplane
    x1 = arange(x1_min-h,x1_max+h,h)
    x2 = hyper(x1, w, b)
    plt.plot(x1, x2)
    plt.xlim(x1_min, x1_max)
    plt.ylim(x2_min, x2_max)
    plt.grid()

    # Show x1 and x2 points according to their classification by the perceptron, i.e. w and b
    x1 = arange(x1_min-h,x1_max+h,h)
    x2 = arange(x2_min-h,x2_max+h,h)
    x1v, x2v = meshgrid(x1, x2, sparse=False, indexing='xy')
    p1 = [(x,y) for (x,y) in zip(x1v.flatten(), x2v.flatten()) if perceptron(array([x,y]), w, b) <= 0]
    plt.scatter(*zip(*p1), marker='o', label='class 0')
    p2 = [(x,y) for (x,y) in zip(x1v.flatten(), x2v.flatten()) if perceptron(array([x,y]), w, b) > 0]
    plt.scatter(*zip(*p2), marker='+', label='class 1')
    plt.xlabel(r'$x_1$', horizontalalignment='right', x=1.0)
    plt.ylabel(r'$x_2$', horizontalalignment='right', y=1.0)
    plt.legend()

    # Hyperplane w
    plt.title(f'$\\theta$={theta:.2f} $w_1$={cos(theta):.2f} $w_2$={sin(theta):.2f}')
    plt.show()
```

```python
def prediction(x, W, b):
    return np.matmul(x,W)+b >= 0

def perceptronStep(X, y, W, b, lr):
    for i in range(len(X)):
        y_hat = prediction(X[i], W, b)

        if y[i] - y_hat == 1:
            W[0] += X[i][0] * lr
            W[1] += X[i][1] * lr
            b += lr
        elif y[i] - y_hat == -1:
            W[0] -= X[i][0] * lr
            W[1] -= X[i][1] * lr
            b -= lr

    return W, b

def trainPerceptron(X, y, lr, n_epochs):
    W = np.array(np.random.rand(2,1))
    b = 2
    lines = []

    for i in range(n_epochs):
        lines.append((-W[0]/W[1], -b/W[1]))
        W, b = perceptronStep(X, y, W, b, lr)

    return lines

res = trainPerceptron(X.T, y, lr=0.1, n_epochs=35)


def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')


for i in range(len(res)):
    plt.scatter(X[0][np.where(y==1)], X[1][np.where(y==1)], color='blue')
    plt.scatter(X[0][np.where(y==0)], X[1][np.where(y==0)], color='red')
    abline(res[i][0], res[i][1])
    plt.show()
```

#### Summary of *The Perceptron Learning Algorithm and its Convergence*

- **Perceptron Learning Algorithm** is a method to train a perceptron to correctly classify data into two categories by adjusting weights incrementally based on errors.

- **Update Rule:** For each misclassified example, the algorithm updates the weights by adding the input vector scaled by the label and a learning rate.

- **Convergence Guarantee:**  
  If the data is linearly separable (there exists a perfect hyperplane separating classes), the algorithm is guaranteed to find weights that perfectly classify all examples in a finite number of steps.

- **Non-separable Data:**  
  If the data is not linearly separable, the perceptron algorithm will not converge but may still find a reasonable separating hyperplane.

- **Margin and Convergence Speed:**  
  The convergence speed depends on the **margin**how well-separated the classes are. Larger margins lead to faster convergence.

- **Geometric Interpretation:**  
  The algorithm iteratively adjusts the weight vector to reduce classification errors by pushing it closer to the ideal decision boundary.

- **Simplicity and Limitations:**  
  While simple and foundational, the perceptron can only solve linearly separable problems. More complex patterns require multilayer networks.

### Sigmoid Neuron

**Perceptron model**, performs binary classification by combining weighted inputs and a bias term through a threshold rule.
- While this model introduced fundamental concepts such as weights, bias, and linear decision boundaries, it is inherently limitedit can only separate data that is **linearly separable** and behaves like a hard **step function**, where even a tiny change in input can abruptly flip the output.

- To address this limitation, we now introduce the **Sigmoid neuron (-neuron)**  the foundational building block of **deep neural networks**.
  - Unlike the perceptron, the sigmoid neuron produces smooth, continuous outputs between 0 and 1, rather than sharp binary values.
  - This smoothness allows small changes in input to produce gradual changes in output, enabling effective learning through gradient-based optimization methods such as **backpropagation**.

- The **transition from a perceptron to a sigmoid neuron** marks a major conceptual leap  **moving from linear**, threshold-based decision boundaries to differentiable,** nonlinear transformations** that allow neural networks to learn complex, non-linear patterns in data.

In the **Sigma (sigmoid) neuron**, the key improvement over the perceptron is replacing the abrupt **step function** output with a **smooth, continuous function**, allowing small changes in the input to produce gradual changes in output. This smooth response enables differentiabilityessential for learning through optimization methods like gradient descent.

**Mathematical Formulation:** The sigmoid (logistic) activation function is defined as:

$$
f(x) = \frac{1}{1 + e^{-(w x + b)}}
$$

where:
- $ w $ = weight parameter associated with the input feature,  
- $ b $ = bias term, and  
- $ x $ = input value.

The function smoothly maps input values from $(-\infty, \infty)$ to the range $(0,1)$. Thus, it behaves like a soft version of the perceptrons step function, allowing probabilistic interpretation:

$$
0 < f(x) < 1 \implies f(x) \text{ can be interpreted as } P(y=1|x)
$$

**Multivariable Extension:**
For multiple inputs $x_1, x_2, x_3, \dots, x_n$, the weighted sum becomes:

$$
z = w_1x_1 + w_2x_2 + w_3x_3 + \dots + w_nx_n + b
$$

and the output (activation) is:

$$
\hat{y} = f(z) = \frac{1}{1 + e^{-z}}
$$


Unlike the perceptrons discrete 0/1 output, the sigmoid neurons curve transitions smoothly between 0 and 1:

$$
\text{for } x \ll 0,\ f(x) \approx 0, \quad \text{for } x \gg 0,\ f(x) \approx 1
$$
This continuous nature enables gradient-based learning.

#### How shape of sigmoid curve changes with parameters `w` and `b`?

Here, $W$ (weight) controls the **steepness** of the curve, while $b$ (bias) controls the **horizontal shift**.

- $W$ determines how sharply the sigmoid transitions from 0 to 1.
- $b$ determines **where** along the x-axis the transition midpoint occurs.

By adjusting $W$ and $b$, we can control both the slope and the position of the logistic curve, allowing it to flexibly fit data across different ranges.

##### Effect of Weight ($W$)

- When $W > 0$: the sigmoid rises smoothly from 0 to 1 as $x$ increases.
- Increasing $W$ makes the transition between 0 and 1 steeper, approaching a step function.
- Decreasing $W$ (e.g., $W < 1$) makes the transition gradual.
- A very small $W$ (close to 0) flattens the curve into an almost straight line.
- A **negative $W$** flips the curve horizontally: output starts near 1 and decreases toward 0.

##### Effect of Bias ($b$)

- The bias $b$ shifts the curve horizontally:
  - Increasing $b$ moves the curve **right**.
  - Decreasing $b$ moves the curve **left**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def Sigmoid(x,w,b):
  return 1/(1+np.exp(-(w*x + b)))
```

#### Differences Between Perceptron and Sigmoid Neuron

- **Output Type:**
  - *Perceptron:* Produces binary output (0 or 1) using a hard threshold (step) function.
  - *Sigmoid Neuron:* Produces continuous output between 0 and 1 using a smooth sigmoid function.

- **Activation Function:**
  - *Perceptron:* Uses a step function, leading to abrupt changes in output.
  - *Sigmoid Neuron:* Uses the sigmoid function $$ \sigma(z) = \frac{1}{1 + e^{-z}} $$, which is differentiable and smooth.

- **Differentiability and Learning:**
  - *Perceptron:* Non-differentiable, limiting learning algorithms to rule-based updates.
  - *Sigmoid Neuron:* Differentiable, enabling gradient-based optimization like backpropagation.

- **Sensitivity to Input Changes:**
  - *Perceptron:* Small changes in input can cause sudden output flips.
  - *Sigmoid Neuron:* Small input changes cause gradual output changes, improving learning stability.

- **Applications:**
  - *Perceptron:* Suitable for simple linear, binary classification tasks.
  - *Sigmoid Neuron:* Suitable for complex networks, supports probabilistic interpretations, and is foundational for deep learning.


### Limitations of Perceptron

Limitations of the perceptron algorithm that motivated significant advances in machine learning:

1. **Limited to linearly seperable data**
    - A major limitation of the perceptron is its inability to handle **non-linearly separable data**.
    - While perceptrons can solve simple binary classification problems like AND or OR where data points are linearly separable, they fail on problems like XOR.
    - The XOR problem needs a decision boundary that is not a single straight lineno linear separator exists to correctly classify all points.
    - To overcome this, **multi-layer perceptrons** (with hidden layers and nonlinear activation functions) are introduced, enabling learning of complex, nonlinear patterns that single-layer perceptrons cannot represent.

2. **No Quality Optimization for the Hyperplane**
    - The perceptron only seeks *some* linear separator for linearly separable data but does **not optimize** the position or quality of the hyperplane.
    - It does not maximize the margin between classes, minimize expected errors, or consider generalization to unseen data.
    - As a result, there could be many valid separators, but the perceptron gives no way to choose the one that is "best" for future predictions or robustness.

3. **Uncertain Convergence and Applicability**
    - The perceptron algorithm only guarantees convergence **if and only if the data is linearly separable**.
    - For non-separable data (which is common in practice), the method may never stop or may loop forever without finding a solution.
    - In such cases, practitioners must arbitrarily limit the number of iterations, lacking a practical convergence criterion or error minimization behavior.


## Feedforward Networks

Basic neural network architecture where signals flow in one direction: from input to output, without cycles.  
  - Can have one or multiple layers with varying connectivity patterns.  
  - Includes single-layer perceptrons and multilayer configurations.

- **Multilayer Perceptrons (MLP):**  
  - A special subset of FNNs characterized by at least one hidden layer and fully connected neurons between layers.  
  - Use nonlinear activation functions in hidden layers enabling modeling of complex, nonlinear decision boundaries.  
  - All MLPs are FNNs, but not all FNNs are MLPs.


#### Mathematical Formulation of Feedforward Neural Networks

A feedforward neural network defines a mapping from an input $\mathbf{x}$ to an output $\mathbf{y}$ via a function $f(\mathbf{x}, \theta)$, where $\theta$ represents the set of learned parameters (weights and biases).

$$
\mathbf{y} = f(\mathbf{x}, \theta)
$$

**Function Approximation**
- Neural networks do not assume the exact form of $f$.
- Instead, they approximate the true function $f^*(\mathbf{x})$ by composing multiple simpler functions, each representing a network layer.

**Layer-wise Composition**
- The mapping $f$ is expressed as a composition of $N$ functions:

$$
f(\mathbf{x}) = f_N (f_{N-1}(\dots f_2(f_1(\mathbf{x}))))
$$

- Each $f_i$ corresponds to a layer transforming its input.
- $N$ is the total number of layers (including hidden and output layers).

**Network Structure**
- The input layer accepts $\mathbf{x}$, which can be a scalar, vector, matrix, or tensor (e.g., images).
- Hidden layers $f_1, \ldots, f_{N-1}$ extract progressively complex features.
- The output layer $f_N$ transforms the last hidden layer output to produce desired $\mathbf{y}$.
- Data flows **forward** from input to output without feedback connections.
- This unidirectional flow motivates the term "feedforward neural network".

## Multilayer (Network of) Peceptron: Stacking layers for better expressivity

A multi-layer perceptron (MLP) is a type of artificial neural network consisting of multiple layers of neurons. The neurons in the MLP typically use nonlinear activation functions, allowing the network to learn complex patterns in data.

- In order to cover a wider range of models, one can stack neurons organized in layers to form a more complex model, such as the model below, which is called a one-hidden-layer model, since an extra layer of neurons is introduced between the inputs and the output:

Multilayer perceptrons (MLPs) address the limitations of single-layer perceptrons by introducing multiple hidden layers that enable the network to learn complex, non-linear patterns.
- Unlike the single-layer perceptron, which can only solve linearly separable problems, MLPs use these hidden layers along with non-linear activation functions to model intricate relationships in data.
- **Feature Abstraction:** Hidden layers in MLPs automatically learn hierarchical feature representations, improving modeling of complicated data structures.
- This makes MLPs powerful deep neural networks capable of handling challenging tasks such as image and speech recognition.
- The MLP architecture consists of an input layer, one or more hidden layers where feature extraction occurs, and an output layer for final predictions.
- **Susceptibility to Overfitting:** With more parameters and layers, MLPs are more prone to overfitting, particularly if the dataset is small relative to the model size.

#### XOR Problem using MLP

**MLP Structure for XOR**

  - **Inputs:** $ X_1 $, $ X_2 $
  - **Hidden Layer:** 2 neurons ($ H_1, H_2 $), step activation
  - **Output Layer:** 1 neuron ($ O $), step activation

**Sample Weights and Biases**

  Lets choose values known to solve XOR:
  - Hidden neuron $H_1$ (AND-like):
    - Weights: $ w_{1,1} = 1 $, $ w_{1,2} = 1 $
    - Bias: $ b_{H1} = -1.5 $
  - Hidden neuron $ H_2 $ (OR-like, then reversed):
    - Weights: $ w_{2,1} = 1 $, $ w_{2,2} = 1 $
    - Bias: $ b_{H2} = -0.5 $

  Output neuron:
  - Weights: $ w_{O,1} = 1 $ (from $ H_1 $), $ w_{O,2} = -2 $ (from $ H_2 $)
  - Bias: $ b_O = 0.5 $

  *Activation function: Step (outputs 1 if input  0, else 0).*

**Manual Calculation: Each Input Combination**

Case 1: $ X_1 = 0, X_2 = 0 $
- $ H_1 = step(1 \times 0 + 1 \times 0 - 1.5) = step(-1.5) = 0 $
- $ H_2 = step(1 \times 0 + 1 \times 0 - 0.5) = step(-0.5) = 0 $
- $ O = step(1 \times 0 + (-2) \times 0 + 0.5) = step(0.5) = 1 $

**Common MLP solution:**

- $ H_1 = step(X_1 + X_2 - 0.5) $ (for OR)
- $ H_2 = step(X_1 + X_2 - 1.5) $ (for AND)

Then, output neuron:
- $ O = step(H_1 - H_2 - 0.5) $ (for XOR)

Let's compute step-by-step:

a. **Input:**
- $ H_1 = step(0 + 0 - 0.5) = step(-0.5) = 0 $
- $ H_2 = step(0 + 0 - 1.5) = step(-1.5) = 0 $
- $ O = step(0 - 0 - 0.5) = step(-0.5) = 0 $

b. **Input:**
- $ H_1 = step(0 + 1 - 0.5) = step(0.5) = 1 $
- $ H_2 = step(0 + 1 - 1.5) = step(-0.5) = 0 $
- $ O = step(1 - 0 - 0.5) = step(0.5) = 1 $

c. **Input:**
- $ H_1 = step(1 + 0 - 0.5) = step(0.5) = 1 $
- $ H_2 = step(1 + 0 - 1.5) = step(-0.5) = 0 $
- $ O = step(1 - 0 - 0.5) = step(0.5) = 1 $

d. **Input:**
- $ H_1 = step(1 + 1 - 0.5) = step(1.5) = 1 $
- $ H_2 = step(1 + 1 - 1.5) = step(0.5) = 1 $
- $ O = step(1 - 1 - 0.5) = step(-0.5) = 0 $

4. **Final XOR Truth Table (MLP Solution)**

## Universal Approximation Theorem

The **Universal Approximation Theorem** is a cornerstone result in the theory of artificial neural networks, **which states that a feedforward neural network with at least one hidden layer containing a finite number of neurons can approximate any continuous function on a compact domain to an arbitrary degree of accuracy, given appropriate parameters and activation functions.**

$$OR$$

**Universal Approximation Theorem: A feed-forward network with a single hidden layer containing a finite number of neurons can approximate continuous functions on compact subsets of $R^n$, under mild assumption on activation function.**

$$OR$$

**The UAT states that a neural network composed of multiple simple functions can approximate any complex function, including those that represent nonlinearly separable data.**

- **Two layers is all that's needed!**:  It states that a neural network with just one hidden layer (often called two layers counting the output layer) can approximate any continuous function to any desired degree of accuracy, provided the hidden layer has enough neurons. The network may need to be very widehaving many neuronsbut theoretically, it can model any function.
  - Importantly, the theorem does not specify how many hidden units are needed to approximate a function, so **some functions may require an impractically large number of units**.

- **Problem - Could be too wide:** By combining many such neurons (blocks), a neural network constructs increasingly accurate approximations of any function. Adding more neurons (blocks) improves the approximation, similar to adding more Lego blocks to build a more detailed structure.

  - However, ***deeper networks can represent complex functions more efficiently than shallow networks*** with very wide layers.
  - Training **deep networks presents computational challenges**; over the past two decades, new algorithms have been developed to make training deep networks more efficient by addressing computational bottlenecks.

- Edges prior to activation are used to build the relationships and edge post activation function are used to join them.

This theorem fundamentally explains why neural networks are so powerful in modeling complex, nonlinear data patterns across a wide range of applications like image recognition, speech processing, and natural language understanding.

- A **single perceptron** serves as a linear classifier. Given an input $ \mathbf{x} $, it outputs either 1 or 0 by essentially determining which side of a hyperplane (a line in 2D, plane in 3D, and higher-dimensional generalization) the input lies:

$$
f(\mathbf{x}) = \text{sign}(\mathbf{w}^T \mathbf{x} + b)
$$

- When you combine multiple perceptrons, each learns to separate the data space using its own linear boundary.

- **Complete representation of data is not possible with single neuron.**

- Using **AND gates**, outputs of several perceptrons can be combined to define a polygonal region covering points belonging to a class. This works because the intersection of half-spaces (defined by perceptrons) forms a polygon.

- An **AND gate** over some perceptrons results in points contained within the polygon formed by those lines:

$$
y_{\text{AND}} = y_1 \land y_2 \land \cdots \land y_k
$$

- Combining these polygons with **OR gates** further enriches the representational power, allowing the network to approximate very intricate, possibly disjoint regions as unions of polygons.

- An **OR gate** combines multiple such polygons:

$$
y_{\text{OR}} = y_{\text{poly}_1} \lor y_{\text{poly}_2} \lor \cdots \lor y_{\text{poly}_m}
$$

This leads to highly expressive decision boundaries capable of fitting complex, intertwined data distributions.

### Why deep Neural networks are needed?

The Universal Approximation Theorem states that a neural network with only two layers (a single hidden layer) and enough neurons can approximate any continuous function to an arbitrary degree of accuracy on a compact domain.

While a two-layer network is theoretically sufficient for universal approximation, in practice deep networks are much more **efficient** and **expressive**. They use their parameters in a way that better captures complex structure in data, which is why modern deep learning architectures often prefer increased depth over extreme width.

#### **Shallow vs. Deep Networks**



- **With the same number of parameters, a deeper network creates more regions in the input space than a shallow one.**
  - For a fixed parameter count, deeper networks can form more complex, highly non-linear partitions and decision boundaries, making them more expressive and capable of modeling more intricate data patterns.
  - In contrast, shallow networks are limited in how many separate "pieces" or regions they can represent, even if they have the same or more parameters.

#### **Depth Efficiency**



- **Depth efficiency** refers to the observation that adding more hidden layers allows a network to approximate complex functions much more efficiently.
  - A deep neural network can achieve the same approximation power as a wide, shallow network but with fewer units per layer. In other words, deep networks can represent certain functions very compactly, while a shallow (single hidden layer) network would need exponentially more neurons to reach the same level of approximation[6].
  - For some tasks, a shallow network might require an infeasible number of hidden units (and, thus, parameters) to match the expressive power of a deeper network of reasonable width.

| Network Structure             | Expressive Power w/ Same Parameters        | Approximation Efficiency                  |
|-------------------------------|--------------------------------------------|-------------------------------------------|
| Shallow (wide, 1 hidden layer)| Fewer regions, less efficient for complex functions | Requires exponentially more units         |
| Deep (more hidden layers)     | More regions, highly expressive            | Achieves same or better approximation with fewer units per layer |

### How a Network Builds Continuous Function?

Each neurons activation (e.g., sigmoid, ReLU) produces nonlinear bumps or step-like responses.
- The first layer maps inputs to these bumps across the input domain.
- The second layer combines (weights and sums) these bumps to approximate the desired function shape.
- To improve approximation accuracy, more neurons can be added, refining the shape.

#### How to Build The Blocks (Neurons)

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-5, 5, 500)
```

```python
# Define two sigmoid functions (neurons) with weights and biases
w1, b1 = np.random.normal(loc = 2, scale = 1) , np.random.uniform(1, 5)
w2, b2 = np.random.normal(loc = 3, scale = 1) , np.random.uniform(1, 5)

sig1 = sigmoid(w1 * x + b1)
sig2 = sigmoid(w2 * x + b2)

# Create a bump by subtracting sig2 from sig1
bump = sig1 - sig2

# Control the height by scaling
height_factor = 2.0
bump_scaled = height_factor * bump

# Plot the building blocks and the bump
plt.figure(figsize=(10, 5))
plt.plot(x, sig1, label='Sigmoid 1')
plt.plot(x, sig2, label='Sigmoid 2')
plt.plot(x, bump, label='Bump (Sigmoid1 - Sigmoid2)')
plt.plot(x, bump_scaled, label='Scaled Bump (Height=2)')
plt.title("Building Blocks (Neurons) as Sigmoid Bumps")
plt.xlabel("Input")
plt.ylabel("Activation")
plt.legend()
plt.grid(True)
plt.show()
```

#### How neural networks use building blocks inside the hidden layer neurons to approximate complex functions?

- Activation functions introduce **nonlinearity** into the neural network. Without them, even multiple layers would merely be equivalent to a single-layer linear model, severely limiting the networks ability to model complex, nonlinear functions.

- For a network to qualify as a **universal function approximator**, its activation function must have certain properties:
  - **Nonlinearity:** The function must be nonlinear to allow networks to combine simple functions into complex ones.
  - **Continuity:** The function should be continuous to ensure smoothness in approximation.
  - **Boundedness (often helpful):** Sigmoid functions saturate at upper and lower limits, which makes stable approximations easier.
  - Certain mild mathematical conditions related to limit behavior at infinity or non-polynomial nature.



**Activation Function Requirements for UAT**

- The original universal approximation theorem was proven for sigmoid functions, relying on their smooth, bounded nonlinear shape.
- Later extensions demonstrated that a wide range of **non-polynomial, continuous activation functions** (including ReLU) also satisfy the approximation conditions.
- The key idea: activation functions provide the **building blocks** and **flexible basis functions** that can be combined by the neural network to approximate any complex function smooth or continuous.

##### Common Activation Functions That Work



1. **Sigmoid (Logistic) Function:**

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

- Outputs values between 0 and 1.
- Smooth, continuous S-shaped curve.
- Historically important and used in original proofs of UAT.
- Allows the neural network to create bump-like building blocks for function approximation.

2. **Hyperbolic Tangent (tanh):**

$$
\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

- Outputs between -1 and 1.
- Zero-centered, which can speed up training compared to sigmoid.
- Has similar approximating power as sigmoid.

3. **Rectified Linear Unit (ReLU):**

$$
\text{ReLU}(x) = \max(0, x)
$$

- Outputs zero for negative inputs and linear for positive inputs.
- Popular in deep learning because of efficient gradient propagation.
- Although piecewise linear and unbounded, ReLU also enables universal approximation with sufficient width/depth.

4. **Leaky ReLU and Variants:**

- Modified ReLU allowing small, non-zero gradients when $$x < 0$$.
- Helps prevent dying ReLU problem and improves training stability.
- Retains universal approximation capabilities.

## Forward Propagation

In artificial neural networks, the information produced by one layer is used as the input to the next layer, forming whats known as a **feedforward neural network**. In this architecture, data always flows in a single directionfrom the input layer, through any hidden layers, and finally to the output layerwithout any cycles or feedback loops. There is no mechanism for information to flow backward or loop within the network.

To understand this clearly, consider how data moves between just two layers: the output of each neuron in the first (input) layer becomes the input to each neuron in the following layer. This **forward-only** flow of information is what distinguishes feedforward networks from other types like recurrent neural networks, where feedback is present.


Forward propagation refers to the process of making predictions in a neural network by passing inputs forward through layers.

- Initially, logistic regression/Perceptron simulates a single neuron. Extending this idea, multiple neurons can process the same inputs differently.

  - **Layer Formation:** Adding more neurons forms a layer (e.g., hidden layer) that extracts features from the inputs.

  - **Deep Networks:** Stacking multiple layers creates deep networks, inspired by the brains multilayered neuron architecture.

  - **Wide and Deep Networks:** The network can grow wider (more neurons per layer) and deeper (more layers), enhancing modeling capacity.

- Neural networks perform a series of feature transformations, gradually extracting complex patterns.
- Initial layers learn simple features (e.g. edges), deeper layers learn complex representations (e.g. faces).
- This hierarchical feature learning underscores the power of deep neural networks.

### Mathematical Model from One Neuron to Multiple Neurons in a Layer

For a single neuron with input vector $ \mathbf{x} $ and weight vector $ \mathbf{w} $, bias $ b $, the output before activation is:

  $$
  z = \mathbf{w}^T \mathbf{x} + b
  $$

- Applying sigmoid activation function, the output is:

  $$
  h = \sigma(z) = \frac{1}{1 + e^{-z}}
  $$


- For $ M $ neurons, each neuron $ j $ computes:

  $$
  h_j = \sigma(\mathbf{w}_j^T \mathbf{x} + b_j)
  $$

- Vectorizing for the whole layer:

  $$
  \mathbf{h} = \sigma(W^T \mathbf{x} + \mathbf{b})
  $$

  where:
  - $ W $ is the weight matrix of shape $ D \times M $
  - $ \mathbf{b} $ is the bias vector of length $ M $

- Activation $ \sigma $ applies element-wise.

**Forward Propagation Between Two Layers**

- Suppose input vector $ \mathbf{X} $ of dimension $ D $ is passed to a layer with $ M $ neurons.

- We represent the weight matrix $ W^{(1)} $ of size $ D \times M $ connecting input to this layer.

- Bias vector $ \mathbf{b}^{(1)} $ has size $ M $.

- The input to neuron $ j $ in the first hidden layer is:

  $$
  z_j^{(1)} = (W^{(1)T} \mathbf{X} + \mathbf{b}^{(1)})_j = \sum_{i=1}^D W_{ij}^{(1)} X_i + b_j^{(1)}
  $$

- Applying an activation function $ \sigma(\cdot) $ element-wise gives hidden layer outputs:

  $$
  h_j^{(1)} = \sigma(z_j^{(1)})
  $$

- Vectorized form:

  $$
  \mathbf{h}^{(1)} = \sigma \bigl( W^{(1)T} \mathbf{X} + \mathbf{b}^{(1)} \bigr)
  $$

**Extending to Multiple Layers**

- For $ L $ layers, with $ \mathbf{z}^{(l)} $ as input to layer $ l $ and $ W^{(l)} $, $ \mathbf{b}^{(l)} $ weights and bias:

  $$
  \mathbf{z}^{(l)} = W^{(l)T} \mathbf{h}^{(l-1)} + \mathbf{b}^{(l)}
  $$

  $$
  \mathbf{h}^{(l)} = \sigma(\mathbf{z}^{(l)})
  $$

- With $ \mathbf{h}^{(0)} = \mathbf{X} $ (input layer).

### Multi-Layer Network Notation

For $ L $ layers, weights and bias have layer indices:

  $$
  h^{(l)} = \sigma(W^{(l)T} h^{(l-1)} + b^{(l)})
  $$

  where $ h^{(0)} = \mathbf{x} $ is input.

- The output at layer $ L $ represents the network prediction.

**Final Layer**

- For binary classification, apply sigmoid to map output into probability $ (0, 1) $.
- For regression, remove final sigmoid for continuous output.

## Backpropagation

- Backpropagation is a key learning algorithm used to train neural networks by minimizing prediction error.
- It adjusts connection weights and biases based on the error between predicted and true outputs.
- This iterative process improves model performance over time.
1. **Forward Pass**
   - Input data flows through the network layer-by-layer producing predictions.
   - Each neuron computes a weighted sum of inputs plus bias and applies an activation function (e.g., sigmoid, ReLU).

2. **Error Calculation**
   - Compute the difference between predicted output $\hat{y}$ and true output $y$ using a loss function, e.g., Mean Squared Error:

  $$
  \text{MSE} = (y - \hat{y})^2
  $$

3. **Backward Pass (Backpropagation)**
   - The error propagates backward through the network.
   - Using the chain rule, gradients of the loss with respect to weights $W$ and biases $b$ are computed for each layer.
   - Gradients indicate how to modify weights to reduce the loss.

4. **Weight Update**
   - Parameters are updated using gradient descent, for example:

  $$
  W^{(t+1)} = W^{(t)} - \eta \frac{\partial \text{Loss}}{\partial W^{(t)}}
  $$

  where $\eta$ is the learning rate.

### Mathematical Formulation

For a layer $l$, given input activations $\mathbf{h}^{(l-1)}$, weights $W^{(l)}$, biases $\mathbf{b}^{(l)}$, and activation function $\sigma$:

  Forward pass:
  $$
  \mathbf{z}^{(l)} = W^{(l)T} \mathbf{h}^{(l-1)} + \mathbf{b}^{(l)}, \quad
  \mathbf{h}^{(l)} = \sigma( \mathbf{z}^{(l)} )
  $$

  Backward pass:
  $$
  \delta^{(L)} = \nabla_{\mathbf{h}^{(L)}} \text{Loss} \odot \sigma'(\mathbf{z}^{(L)})
  $$

  and for preceding layers:

  $$
  \delta^{(l)} = (W^{(l+1)} \delta^{(l+1)}) \odot \sigma'(\mathbf{z}^{(l)})
  $$

  where $\delta^{(l)}$ is the error term at layer $l$, and $\odot$ denotes element-wise multiplication.

- Gradients for weight and bias updates are:

  $$
  \nabla_{W^{(l)}} = \mathbf{h}^{(l-1)} \delta^{(l)T}, \quad \nabla_{\mathbf{b}^{(l)}} = \delta^{(l)}
  $$



#### Training an MLP on XOR Problem

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt

# XOR dataset
gX = np.array([[0,0],[0,1],[1,0],[1,1]])
gY = np.array([0,1,1,0])

# Build model: 2 inputs, one hidden layer with 2 neurons, output layer with 1 neuron
model = Sequential()
model.add(Dense(2, input_dim=2, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

# Compile model with SGD optimizer and binary cross-entropy loss
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])

# Train model for 1000 epochs
history = model.fit(gX, gY, epochs=1000, verbose=0)

# Plot training accuracy
plt.plot(history.history['accuracy'])
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()

# Evaluate model
loss, accuracy = model.evaluate(gX, gY, verbose=0)
print(f'Training accuracy: {accuracy*100:.2f}%')
```

## Neurons and Neural Networks

A **neural network** is a computational model inspired by the human brain, composed of many interconnected processing units called **artificial neurons**. These neurons work collectively to solve complex problems such as classification, regression, and pattern recognition. Understanding how individual neurons function and how they are organized gives insight into the power of neural networks.

### The Artificial Neuron



An **artificial neuron** mimics some aspects of biological neurons. It receives multiple inputs, processes them by computing their weighted sum, and then passes this sum through an **activation function** to produce an output.

- The input vector: $ \mathbf{x} = [x_1, x_2, ..., x_n] $
- The weight vector: $ \mathbf{w} = [w_1, w_2, ..., w_n] $
- A bias term: $ b $

The weighted sum (also called the net input) is:

$$
z = \mathbf{w}^T \mathbf{x} + b = \sum_{i=1}^{n} w_i x_i + b
$$

The output of the neuron is obtained by applying an **activation function** $ \phi(z) $:

$$
output = \phi(z)
$$

### Difference Between a Perceptron and an Artificial Neuron

- A **perceptron** is the simplest neuron model introduced by Rosenblatt. It calculates a weighted sum of inputs and passes it through a **step function**, which outputs a binary signal (usually 0 or 1) depending on whether the sum crosses a threshold.
  
- An **artificial neuron** generalizes this idea by allowing the activation function $ \phi $ to be any **non-linear function** rather than just the step function. Examples of activation functions include sigmoid, tanh, and ReLU.

This nonlinearity is critical because it allows neural networks to learn complex, nonlinear mappings from inputs to outputs, a capability not possible with just perceptrons.

### Neural Network Architecture

Neurons in a network are arranged in **layers**:

- **Input Layer:**  
  Contains as many neurons as the number of input features in the dataset. It feeds the data into the network but does not perform computations itself.

- **Hidden Layers:**  
  One or more layers between the input and output that perform computations. Neurons here apply weighted sums followed by activation functions, extracting progressively complex features.

- **Output Layer:**  
  For classification tasks, the number of output neurons generally corresponds to the number of classes. For regression tasks, typically only one neuron outputs the numeric prediction.

### Specifications of a Neural Network

To fully specify a neural network, one needs to define:

1. **Network Topology (Structure):**
   - The number of layers, neurons per layer.
   
2. **Graph**
  - *Nodes (Neurons):*
    - Computational units that receive inputs and produce outputs.
  - *Edges (Connections):*
    - Connections between neurons, each having an associated weight, representing the strength of the connection.
   
3. **Input Layer:**
   - Number of neurons equal to the number of input features.
   
4. **Output Layer:**
   - Number of neurons equal to the number of output variables or classes.
   
5. **Weights ($ \mathbf{w}_i $):**
   - Parameters that modulate input signals. Each connection has a weight.
   
6. **Biases:**
   - Additional parameter added to weighted sums to allow shifting and better fit.
   
7. **Activation Functions:**
   - Non-linear functions applied to weighted sums (e.g., sigmoid, ReLU).

## Inputs and Outputs of a Neural Network

Neural networks are structured as collections of artificial neurons arranged in layers. Understanding how inputs enter the network and how outputs are produced is fundamental for designing and applying neural networks effectively.

### Inputs to a Neural Network

- **Number of Neurons in the Input Layer:**  
  Determined by the number of features or attributes in the dataset. Each neuron corresponds to one numeric input feature.

- **Types of Input Data:**  
  Neural networks accept only **numeric inputs**. Non-numeric data such as text, speech, or images must be converted into numeric form before feeding into the network.

- **Input Formats:**

  1. **Numeric Data:**  
     Directly used as inputs once normalized or standardized.

  2. **Text Data:**  
     - Represented as **one-hot vectors**: For a vocabulary $ V $, each word $ w_n $ is encoded as a vector of length $|V|$ with one element set to 1 and zeros elsewhere.  
     - **Word embeddings:** Dense, lower-dimensional continuous vector representations (e.g., Word2Vec, GloVe) that capture semantic relationships and reduce dimensionality.

  3. **Image Data:**  
     Images are naturally numeric arrays of pixel values.  
     - A grayscale (black-and-white) image of size $$H \times W$$ is represented as an $H \times W$ matrix of pixel intensities, flattened into a vector with $ H \times W $ input neurons.  
     - For colored images (RGB), each pixel has three channels (Red, Green, Blue), so a $ H \times W $ image is represented as an $ H \times W \times 3 $ tensor, corresponding to that many input neurons.

### Example

- The MNIST dataset consists of black-and-white images of handwritten digits sized $28 \times 28$.  
- Each image is flattened into a vector of size 784 and fed into the input layer of the network.

### Outputs of a Neural Network

- **Output Layer Size:**

  - **Classification Tasks:**  
    The number of output neurons equals the number of target classes.  
    For example, MNIST digit classification has 10 output neurons representing digits 0 through 9.

  - **Regression Tasks:**  
    Typically, a single output neuron predicts a continuous numeric value.

- **Output Activation:**

  - **Softmax Activation:**  
    Used for multi-class classification, converts raw scores into probabilities $ p_i $ for each class $ i $, with the sum of all $ p_i = 1 $.

  - **Sigmoid Activation:**  
    Used for binary classification, outputs the probability of the positive class.

**Relationship Between Softmax and Sigmoid**

- The sigmoid function is a special case of softmax when there are only two classes.

$$
p_0 = \frac{e^{z_0}}{e^{z_0} + e^{z_1}}, \quad p_1 = \frac{e^{z_1}}{e^{z_0} + e^{z_1}}
$$

If you divide numerator and denominator as shown, you derive the sigmoid formulation from the softmax.



## Simplifying Assumptions in Neural Networks

Neural networks, especially large ones, can have very complex and intricate architectures. To analyze, understand, and implement these networks effectively, standard simplifying assumptions are often made about their structure and information flow. These assumptions not only make the networks mathematically tractable but also help standardize design and training practices in machine learning.




1. **Layered Topology:**  
   Neurons are organized into discrete layers. These layers are sequentially stacked. The information flows from the input layer through one or more hidden layers and finally to the output layer. This assumption reflects the feedforward architecture common in many neural networks.

2. **No Intra-layer Connections:**  
   Neurons within the same layer do **not** interact or send signals to each other. Communication happens only between neurons in adjacent layers.

3. **Clear Input and Output Boundaries:**  
   All the input features enter only via the **input layer**, and all final outputs emerge through the **output layer**. There are no shortcuts or skip connections bypassing these layers (note: this is an assumption that simplifies many architectures but is relaxed in advanced models like ResNets).

4. **Dense Layer Connections:**  
   Each neuron in layer $l$ is connected to every neuron in the next layer $l+1$. This "fully connected" assumption ensures maximum information flow and is typical of classical Multi-Layer Perceptrons (MLPs).

5. **Weights on Connections:**  
   Every edge connecting two neurons carries a **weight** $w$ representing the strength of that connection, dictating how much the input from the preceding neuron influences the next.

6. **Bias Terms:**  
   Each neuron has an associated **bias** parameter $b$ which allows shifting the activation function and helps increase the models flexibility.

7. **Uniform Activation Functions:**  
   It is usually assumed that all neurons within a given layer use the **same activation function** (e.g., ReLU, sigmoid). This provides consistency and simplifies training. Different layers may use different activations, but within a layer they are uniform.

**Summary of Neural Network Specifications**

To build or analyze a neural network, one specifies:

- **Network Topology:** Number of layers, size of each layer, connectivity pattern.
- **Input Layer:** Number of neurons = number of input features.
- **Output Layer:** Number of neurons depends on the problem (number of classes for classification, or 1 for regression).
- **Weights:** All the parameters of connections between neurons.
- **Biases:** One bias per neuron except input neurons.
- **Activation Functions:** Nonlinear functions applied at each neuron's output.


### Neural Network Training

Neural network training is the process of finding the best values for the weights and biases that connect the neurons, so the network can make accurate predictions. Like models such as linear or logistic regression, neural networks are optimized to minimize a chosen loss function by adjusting these parameters.

- **Parameters learned:**  
  During training, the network automatically learns the optimal values of **weights and biases** through data. This process is algorithmic and data-driven.

- **Hyperparameters set by the user:**  
  The network architect (you) chooses hyperparameters such as the number of layers, units in each layer, activation functions, batch size, and sometimes the learning ratethese determine the networks structure and behavior before training begins.

- **Training Process Overview:**
  1. **Fix the topology:** Decide on the number of layers, neurons per layer, activation functions, and input/output shape.
  2. **Initialize parameters:** Set initial (often random) values for weights and biases.
  3. **Forward pass:** For each batch, compute outputs and compare predictions to ground truth using a loss function.
  4. **Backward pass (Backpropagation):** Compute gradients of the loss with respect to each parameter, indicating how to adjust weights and biases to reduce loss.
  5. **Parameter update:** Update weights and biases, typically using an optimizer like gradient descent.
  6. **Iterate:** Repeat the process for multiple epochs over the training data until convergence or stopping criteria are met.

| Aspect                  | Who decides?   | Examples                                               |
|-------------------------|----------------|--------------------------------------------------------|
| Parameters (learned)    | Algorithm      | Weights, biases (updated during training)              |
| Hyperparameters (fixed) | User/designer  | Layer count, neuron count, activation, batch size, etc.|
